**COLUMN DESCRIPTION**

Instrumentalness: This value represents the amount of vocals in the song. The closer it is to 1.0, the more instrumental the song is.
Acousticness: This value describes how acoustic a song is. A score of 1.0 means the song is most likely to be an acoustic one.

Liveness: This value describes the probability that the song was recorded with a live audience. According to the official documentation “a value above 0.8 provides strong likelihood that the track is live”.

Speechiness: “Speechiness detects the presence of spoken words in a track”. If the speechiness of a song is above 0.66, it is probably made of spoken words, a score between 0.33 and 0.66 is a song that may contain both music and words, and a score below 0.33 means the song does not have any speech.

Energy: “(energy) represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy”.

Danceability: “Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable”.

Valence: “A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry)”.

```python
#convert duration_ms into hour:minute:seg format

from datetime import datetime

def duration_min_seg(time_ms):

  if time_ms<900:
    return "00:00:00"

  elif time_ms>=1000 and time_ms<60000:
      seg=time_ms/1000
      seg=int(seg)
      res=0
      hour=0
      min=0


  elif time_ms>=60000 and time_ms<=36000000:
      seg=time_ms/1000
      hour=0
      min=int(seg/60)
      seg%=60
      seg=int(seg)


  else:
      min=0
      hour=int(seg/3600)
      res=3600*hour
      min1=seg-res
      if min1>=60:
        min=min1/60
        min=int(min)
        seg%=60
        seg=int(seg)

  return "{0}:{1}:{2}".format(hour,min,seg)



#res=duration_min_seg(1000000)
#print(res)

df['duration_min_seg']=df['duration_ms'].map(lambda x:duration_min_seg(x))
#print(df['duration_min_seg'])


#convert to time function

def convert_time(dt_string):

  return format(datetime.strptime(dt_string, "%H:%M:%S"),"%H:%M:%S")

df['duration_min_seg']=df['duration_min_seg'].map(lambda x:convert_time(x))

#print(df['duration_min_seg'])

df_new=df.drop(columns=['playlist_url','duration_ms','artist_id','time_signature','tempo','key','valence','acousticness','loudness','instrumentalness'])

#df_new.info()
df_new.reset_index(drop=True)

```

```python

#Speechiness: “Speechiness detects the presence of spoken words in a track”. If the speechiness of a song is above 0.66, it is probably made of spoken words,
#a score between 0.33 and 0.66 is a song that may contain both music and words, and a score below 0.33 means the song does not have any speech.

#>0.66-set label "spoken words"
#0.33-0.66- set label "music and words"
#<0.33-set label "no speech"

def label(row):
  if row.speechiness<0.33:
    return "spoken words"
  elif row.speechiness>=0.33 and row.speechiness<=0.66:
    return "music and words"
  else:
    return "no speech"

df_new['speechiness']=df.apply(label,axis=1)
df_new
```

```python
#Transforming danceability to labels.

#Danceability: “Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo,
#rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable”.


#<0.5 set label: "not danceable"
#0.6-0.8 - set label: "mid danceable"
#>0.8 -set label: "danceable"


def danceable(row):
  if row.es_bailable<=0.5:
    return "not_danceable"
  elif row.es_bailable>0.5 and row.es_bailable<=0.7:
    return "mid_danceable"
  else:
    return "super_danceable"

df_renamed['es_bailable']=df_renamed.apply(danceable,axis=1)

df_renamed[df_renamed.loc[:,'es_bailable']=="super_danceable"]

```

```python
#The same for energy
#energy
def energetic(row):
  if row.energía<=0.5:
    return "not_energetic"
  elif row.energía>0.5 and row.energía<=0.8:
    return "mid_energetic"
  else:
    return "super_energetic"

df_renamed['energía']=df_renamed.apply(energetic,axis=1)
df_renamed
```

```python
#Descriptor count

is_danceable_count=df_renamed.es_bailable.map(lambda desc:"super_danceable" in desc ).sum()
is_energetic_count=df_renamed.energía.map(lambda desc:"super_energetic" in desc).sum()
is_mid_danceable_count=df_renamed.es_bailable.map(lambda desc:"mid_danceable" in desc ).sum()
is_mid_energetic_count=df_renamed.energía.map(lambda desc:"mid_energetic" in desc).sum()
is_not_danceable_count=df_renamed.es_bailable.map(lambda desc:"not_danceable" in desc ).sum()
is_not_energetic_count=df_renamed.energía.map(lambda desc:"not_energetic" in desc).sum()

descriptor_counts=pd.Series([is_danceable_count,is_mid_danceable_count,is_not_danceable_count,is_energetic_count,is_mid_energetic_count,is_not_energetic_count], index=["super_danceable","mid_danceable","not_danceable","super_energetic","mid_energetic","not_energetic"])
descriptor_counts
```

```python
#Concatenate 10 iterations of data

import time

dict_datos={}

for i in range(0,10):

  try:
    time.sleep(5)
    r=requests.get(endpoint)

    if r.status_code==200:
      b=r.json()

      data=b['list']

      dict_datos[i]=data

  except:
      continue
```

```python
for d in dict_datos:
  print(d)
```

```python
for dat in dict_datos:
  normalized=pd.json_normalize(dict_datos[dat])
  df=pd.DataFrame.from_dict(normalized)
  dict_datos[dat]=df
```

```python
#Transform data to dataframe
lista_dataframes=[]

for key in dict_datos:
  lista_dataframes.append(dict_datos[key])
```

```python
#save data to csv file
df_complete=pd.concat(lista_dataframes,axis=0).reset_index(drop=True)
df_complete.to_csv("/content/gdrive/MyDrive/dataset_bedu proyecto/weather_forecast.csv")
```
